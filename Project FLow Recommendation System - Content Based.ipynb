```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        word = ps.stem(i)
        y.append(word)
    return " ".join(y)
```

```python
movies_data['tags'].apply(stem)
```

```python
## calculate similarity between text 1 and text 2
## convert text -> vector

## convert tags -> vector , closest vectors will have the highest similarity -> text vectorisation
```

```python
## Using bag of words
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_data['tags']).toarray()
```

```python
cv.vocabulary_
```

```python
## calculate distance of each movie with another
## more distance less similar-> use cosine similarity , angle between them -> high dimensional , 
## eucledian distance will fail for higher dimensions
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity ## similarity of each movie with another movie

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:5]
```

```python
def recommend(movie):
    movie_index = movies_data[movies_data['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
    for i in movies_list:
        print(movies_data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
